# Data Processing


### Importing libraies

```python
import numpy as np
import pandas as pd
import tensorflow as tf
```

###Importing the dataset

```python
dataset = pd.read_csv('Dataset_master.xlsx - American Express User Exit Pred.csv')
X = dataset.iloc[:, 0:-1].values
Y = dataset.iloc[:, -1].values
```

### Encoding Catagorical Data

####LabelEncoding - Dependent Variable (Gender Column)

```python

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
X[:,2] = le.fit_transform(X[:,2])
```

#### OneHotEncoding - Independent Variable (Geography Column)

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [1])], remainder='passthrough')
X = np.array(ct.fit_transform(X))
```

### Spliting the data into testset and traing set

```python
from sklearn.model_selection import train_test_split
X_train, X_test , Y_train , Y_test = train_test_split(X, Y, test_size =0.2 , random_state = 0)
```

```python
print(Y_train)
```

###Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

##ANN Implimentation


###Initialisation

```python
ann = tf.keras.models.Sequential()
```

###Adding Input Layer & first hidden layer

```python
ann.add(tf.keras.layers.Dense(units = 6, activation = 'relu'))
```

###Adding Second Hidden Layer

```python
ann.add(tf.keras.layers.Dense(units = 6, activation = 'relu'))
```

###Adding Output Layer

```python
ann.add(tf.keras.layers.Dense(units = 1, activation = 'sigmoid'))
```

###ANN Training

###Compiling ANN

```python
ann.compile(optimizer = 'adam' , loss = 'binary_crossentropy' , metrics = ['accuracy'])
```

### Training on Training Dataset

```python
print(X_train)
```

```python
ann.fit(X_train,Y_train,batch_size=32,epochs=120)
```

## Predictions

###Single Prediction

```python
print(ann.predict(sc.transform([[0.0, 1.0, 0.0, 501, 0, 32, 2, 0.0, 4, 1, 545501]])) > 0.5)
```

###Prediction on testset

```python
y_pred = ann.predict(X_test)
y_pred = (y_pred > 0.5)
print(np.concatenate((y_pred.reshape(len(y_pred), 1), Y_test.reshape(len(Y_test), 1)), 1))
```

###Confusion Matrix

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(Y_test, y_pred)
print(cm)
accuracy_score(Y_test, y_pred)
```
